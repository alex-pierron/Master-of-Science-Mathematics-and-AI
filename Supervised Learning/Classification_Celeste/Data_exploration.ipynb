```python
num_rows, num_cols = 4, 4  
fig_width, fig_height = 12, 12

fig, axes = plt.subplots(num_rows, num_cols, figsize=(fig_width, fig_height))

axes = axes.ravel()

for i, column in enumerate(data_train.columns):
    ax = axes[i]
    ax.boxplot(data_train[column])
    ax.set_title(column)
    ax.set_xlabel("Value")
    ax.set_ylabel(column)

for i in range(len(data_train.columns), num_rows * num_cols):
    fig.delaxes(axes[i])

plt.tight_layout()

plt.show()
```
