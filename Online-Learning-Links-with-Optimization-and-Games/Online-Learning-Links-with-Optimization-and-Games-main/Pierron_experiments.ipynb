# I°/ Games used

The following games are stored in the "zero_sum_games" dictionnary accessible in the script file:
1. **Rock-Paper-Scissors**:
   - **Purpose**: Classic game to introduce strategic decision-making.
   - **Selection Reason**: Simple and widely known game.

2. **Modified Rock-Paper-Scissors**:
   - **Purpose**: Extends original game with additional choices.
   - **Selection Reason**: Shows variations in game rules and strategies.

3. **Trading Game**:
   - **Purpose**: Simulates decision-making in financial markets.
   - **Selection Reason**: Gives a more complex game with different values that 0 or 1 in the payoff matrix. Asymmetric scenario.

4. **Guess the Answer**:
   - **Purpose**: Highlights decision-making under uncertainty.
   - **Selection Reason**: Illustrates asymmetric scenario with a slightly larger decision space than the trading game and different values.

5. **Stock Market Duel**:
   - **Purpose**: Provides a more complex market decision-making environment.
   - **Selection Reason**: Gives the most complex game available in this study with different values than 0 or 1 in the payoff matrix. Asymmetric scenario.

---

This last game is callable through the function "generate_random_payoff_matrix" (accessible in the script file) and is not integrated inside the "zero_sum_games" dictionnary.

**Random Game (Callable Function)**:
- **Purpose**: Allows generation of a random game with a fixed size for the payoff matrix.
- **Selection Reason**: Offers flexibility to explore various game scenarios by generating custom-sized payoff matrices.


```python
random_game_payoff_matrix = random_payoff_matrix(-25, 50, 17, 31, dtype=int, seed=42)
```

# II°/ Extrapolation Algorithm

In this part we perform numerical experiments to determine for each game the best hyperparameters when using the extrapolation algorithm defined in question V. For each game, we test for the proposed value of gamma and several other to compare the  behavior of the algorithm. \
Entering manually the proposed value of gamma in the report is not necessary and is directly done by default for the grid_search function.\
The ouputs display which parameters are the best among those tested.

### II.A°/ Rock-Paper-Scissors

```python
parameter_grid = {'gamma': [0.001, 0.01, 0.1, 1, 10], 'max_iter': [10000]}
results_extra_rock_paper = grid_search(zero_sum_games["Rock-Paper-Scissors"]["payoff_matrix"], parameter_grid,epsilon=1e-4)
```

### II.B°/ Modified Rock-Paper-Scissors

```python
parameter_grid = {'gamma': [0.001, 0.01, 0.1, 1, 10], 'max_iter': [10000]}
results_extra_modified_rock_paper = grid_search(zero_sum_games["Modified Rock-Paper-Scissors"]["payoff_matrix"],
                                                parameter_grid,
                                                epsilon=1e-4)
```

### II.D°/ Guess the answer

```python
parameter_grid = {'gamma': [0.001, 0.01, 0.1, 1, 10], 'max_iter': [10000]}
results_extra_guess_answer = grid_search(zero_sum_games["Guess the answer"]["payoff_matrix"],
                                                parameter_grid,
                                                epsilon=1e-4)
```

### II.E°/ Stock Market Duel

```python
parameter_grid = {'gamma': [0.001, 0.01, 0.1, 1, 10], 'max_iter': [10000]}
results_extra_stock_duel = grid_search(zero_sum_games["Stock Market Duel"]["payoff_matrix"],
                                                parameter_grid,
                                                epsilon=1e-4)
```
